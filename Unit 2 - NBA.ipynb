---
### How many games need to win to make it to the playoffs?

```r
table(NBA$W, NBA$Playoffs)
```

Around 42 is the threshold of for high certainity. Around 35 is low certainity.

---
Can we see the diff between points scored and points allowed in the regular season to predict if the team will win?




```r
NBA$PTSdiff = NBA$PTS - NBA$oppPTS
plot(NBA$PTSdiff, NBA$W)
```

There is a strong coorelation here so linear regression is a good fit.


```r
WinsReg = lm(W~PTSdiff, data= NBA )
summary(WinsReg)
```

### Strong r2  very strong linear relationship

---
$W=41+0.0326 * PTSdiff$

41 From intercept coe/b
.0326 from PTSdiff coe/mx

If wins was 42, what needs to be the PTSdiff?

$ PTSdiff \geq 42-41/ 0.0326 = 30.67 $

Calculate points scored

```r
PointsReg = lm(PTS ~ X2PA + X3PA + FTA + AST + ORB + DRB + TOV + STL + BLK, data=NBA)
summary(PointsReg)
```

Compute Residuals


```r
PointsReg$residuals
```

```r
SSE = sum((PointsReg$residuals)^2)
SSE
```

```r
RMSE=sqrt(SSE/nrow(NBA))
RMSE
```

```r
mean(NBA$PTS)
```

On average we make 184.404 per point compare to mean/average.

Feature selection to optimize the model

We choose TOV because it has a high p value or means that is the least stats sig variable.



```r
PointsReg2 = lm(PTS ~ X2PA + X3PA + FTA + AST + ORB + DRB  + STL + BLK, data=NBA)
summary(PointsReg2)

```

```r
PointsReg3 = lm(PTS ~ X2PA + X3PA + FTA + AST + ORB + STL + BLK, data=NBA)
summary(PointsReg3)

```

```r
PointsReg4 = lm(PTS ~ X2PA + X3PA + FTA + AST + ORB + STL, data=NBA)
summary(PointsReg4)

```

```r
SSE_4 = sum((PointsReg4$residuals)^2)
SSE_4

```

```r
RMSE_4=sqrt(SSE_4/nrow(NBA))
RMSE_4
```

---

```r
NBA_test = read.csv("NBA_test.csv")
PointsPrediction = predict(PointsReg4, newdata=NBA_test)
SSE = sum((PointsPrediction - NBA_test$PTS) ^2)
SST = sum((mean(NBA$PTS)- NBA_test$PTS)^2)
R2=1-SSE/SST
R2
```

```r
RMSE = sqrt(SSE/nrow(NBA_test))
RMSE
```
